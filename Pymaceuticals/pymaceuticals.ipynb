```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug = "data/mouse_drug_data.csv"
clinical = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
clinical=pd.read_csv(clinical)
mouse_data=pd.read_csv(mouse_drug)

# Combine the data into a single dataset
combined_pymaceuticals=pd.merge(clinical, mouse_data, on=['Mouse ID', 'Mouse ID'], how='left')

# Display the data table for preview
combined_pymaceuticals.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_volume=combined_pymaceuticals.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"]
mean_tumor_volume= tumor_volume.mean()
# Convert to DataFrame
mean_tumor_volumedf=mean_tumor_volume.reset_index()
# Preview DataFrame
mean_tumor_volumedf.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
se_tumor_volumes=tumor_volume.sem()
# Convert to DataFrame
se_tumor_volumesdf=se_tumor_volumes.reset_index()
# Preview DataFrame
se_tumor_volumesdf.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
grouped_df=pd.merge(mean_tumor_volumedf, se_tumor_volumesdf, how="left", on=["Drug", "Timepoint", "Tumor Volume (mm3)"])
grouped_pivot=grouped_df.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")
# Preview that Reformatting worked
grouped_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis=np.arange(0,50,5)
marker_list=['.', ',', 'o', 'v', '^', 's', 'p', '*','D','d']
fig1=plt.figure(figsize=(8,5))
ax1 = fig1.add_subplot(111)

i=0
for drug in grouped_pivot.columns:
    ax1.errorbar(x_axis,mean_tumor_volume[drug],yerr=se_tumor_volumes[drug],fmt=marker_list[i],label=drug, linestyle="dashed", linewidth=0.50)
    i+=1
plt.legend(loc='upper left')
plt.title('Tumor Response to Treatment',fontsize=14,fontweight='bold')
plt.xlim(min(x_axis)-5,max(x_axis)+5)
plt.xticks(x_axis)
plt.xlabel('Time (Days)',fontsize=12)
plt.ylabel('Tumor Volume(mm3)',fontsize=12)
plt.grid(linestyle='dotted')
# Save the Figure
fig1.savefig("analysis/Fig1.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_sites_mean=combined_pymaceuticals.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]
# Convert to DataFrame
met_sites_mean=pd.DataFrame(met_sites_mean)
# Preview DataFrame
met_sites_mean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sites_sem=combined_pymaceuticals.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame
met_sites_sem=pd.DataFrame(met_sites_sem)

# Preview DataFrame
met_sites_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_sites_mean=met_sites_mean.reset_index()
met_sites_pivot_mean=met_sites_mean.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

met_sites_sem=met_sites_sem.reset_index()
met_sites_pivot_sem=met_sites_sem.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

# Preview that Reformatting worked
met_sites_pivot_mean
```

```python
# Generate the Plot (with Error Bars)
fig2=plt.figure(figsize=(8,4.5))
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean['Capomulin'], yerr=met_sites_pivot_sem['Capomulin'],
            color="r", marker='o', markersize=5, linestyle="dashed", linewidth=0.50, label='Capomulin')
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean['Infubinol'], yerr=met_sites_pivot_sem['Infubinol'],
            color="b", marker='^', markersize=5, linestyle="dashed", linewidth=0.50, label='Infubinol')
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean['Ketapril'], yerr=met_sites_pivot_sem['Ketapril'],
            color="g", marker='s', markersize=5, linestyle="dashed", linewidth=0.50, label='Ketapril')
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean['Placebo'], yerr=met_sites_pivot_sem['Placebo'],
            color="k", marker='d', markersize=5, linestyle="dashed", linewidth=0.50, label='Placebo')

plt.title('Metastatic Spread During Treatment',fontsize=12)
plt.xlim(min(x_axis)-2,max(x_axis)+2)
plt.xticks(x_axis)
plt.xlabel('Treatment Duration (Days)',fontsize=12)
plt.ylabel('Met Sites',fontsize=10)
plt.grid(linestyle='dotted')
plt.legend(loc="best", fontsize="small", fancybox=True)
# Save the Figure
fig2.savefig("analysis/Fig2.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_rates=combined_pymaceuticals.groupby(["Drug", "Timepoint"]).count()["Tumor Volume (mm3)"]

# Convert to DataFrame
survival_rates=pd.DataFrame(survival_rates)
survival_ratesdf=survival_rates.rename(columns={"Tumor Volume (mm3)":"Mouse Count"})

# Preview DataFrame
survival_ratesdf.head().reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_ratesdf=survival_ratesdf.reset_index()
survival_rates_pivot=survival_ratesdf.pivot(index="Timepoint", columns="Drug")["Mouse Count"]
# met_sites_pivot_sem=met_sites_sem.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

# Preview the Data Frame
survival_rates_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
fig3=plt.figure(figsize=(8,4))
plt.plot(100 * survival_rates_pivot["Capomulin"]/25, color="r", marker='o', markersize=5, linestyle="dashed", linewidth=0.50, label='Capomulin')
plt.plot(100 * survival_rates_pivot["Infubinol"]/25, color="b", marker='^', markersize=5, linestyle="dashed", linewidth=0.50, label='Infubinol')
plt.plot(100 * survival_rates_pivot["Ketapril"]/25, color="g", marker='s', markersize=5, linestyle="dashed", linewidth=0.50, label='Ketapril')
plt.plot(100 * survival_rates_pivot["Placebo"]/25, color="k", marker='d', markersize=5, linestyle="dashed", linewidth=0.50, label='Placebo')

plt.title('Survival During Treatment',fontsize=12)
plt.xlim(min(x_axis)-2,max(x_axis)+2)
plt.xticks(x_axis)
plt.xlabel('Time (Days)',fontsize=12)
plt.ylabel('Survival Rate (%)',fontsize=10)
plt.grid(linestyle='dotted')
plt.legend(loc="best", fontsize="small", fancybox=True)
# Save the Figure
fig3.savefig("analysis/Fig3.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
#loc first and last rate to calcualte percentage change of tumor volume
for drug in grouped_pivot:
    firsttumorvolume=grouped_pivot.iloc[0]
    lasttumorvolume=grouped_pivot.iloc[-1]
change=lasttumorvolume-firsttumorvolume
percentagechange=(change/firsttumorvolume)*100
# Display the data to confirm
percentagechange              
```

```python
# Store all Relevant Percent Changes into a Tuple
pct_changes= (percentagechange ["Capomulin"], 
              percentagechange ["Infubinol"],
              percentagechange["Ketapril"],
              percentagechange ["Placebo"])
# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
ind = np.arange(len(pct_changes))
width=1
rectsPass = ax.bar(ind[0], pct_changes[0], width, color='green')
rectsFail = ax.bar(ind[1:], pct_changes[1:], width, color='red')

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_ylim([-30,70])
ax.grid(True)

# Use functions to label the percentages of changes
def autolabelFail(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 3,
                '%d%%' % int(height),
                ha='center', va='bottom', color="white")

def autolabelPass(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., -8,
                '-%d%%' % int(height),
                ha='center', va='bottom', color="white")

# Call functions to implement the function calls
autolabelPass(rectsPass)
autolabelFail(rectsFail)

# Save the Figure
fig.savefig("analysis/Fig4.png")

# Show the Figure
fig.show()
```
